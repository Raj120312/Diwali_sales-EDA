```python
import numpy as np
```

```python
import pandas as pd
```

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
df = pd.read_csv('Diwali Sales Data.csv',encoding='ISO-8859-1')
```

```python
df.shape
```

```python
df.head()
```

```python
df.info()
```

```python
df.drop(['Status','unnamed1'],axis=1,inplace=True) # dropping status and unnamed1 which are unnecessary column
```

```python
df.info()
```

```python
pd.isnull(df).sum() 
```

```python
df.dropna(inplace=True) #dropping null rows
```

```python
df.info()
```

```python
df['Amount'] = df['Amount'].astype('int')
```

```python
df['Amount']
```

```python
df['Amount'].dtypes
```

```python
df.columns
```

```python
df.rename(columns={'Marital_Status':'Marriage'},inplace=True)
```

```python
df.columns
```

```python
df.head()
```

```python
df.describe()
```

```python
ax = sns.countplot(x='Gender',data =df) # a normal countplot.
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# from the above graph we can see that count of females are more in the dataset
```

```python
sales_gen=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False) #in this we are grouping the the dataframe based on gender and sorting the gender based on sum of the amount spent
```

```python
sns.barplot(x = 'Gender',y = 'Amount' , data = sales_gen)
```

```python
ax = sns.countplot(data = df , x = 'Age Group' , hue = 'Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#from the above graph , we can see that the count of age group of 26-35 is highest
```

```python
sales_age=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Age Group',y = 'Amount', data = sales_age)
```

```python
#from the above graph , we can also see that the age group 26-35 spend most
```

```python
plt.figure(figsize=(20,10))
ax = sns.countplot(x ='State',data = df,hue="State",palette="flare")
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

```python
#from the above we can see that Uttar Pradesh has most sales while Telangana state has least sales 
```

```python
sales_state=df.groupby(['State'],as_index= False)['Orders'].sum().sort_values(by='Orders',ascending= False).head(8)
ax = sns.barplot(x='State',y='Orders',data =sales_state,hue="State",palette="flare")
sns.set(rc={'figure.figsize':(20,10)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#from the above figure , we can see that most orders came from Uttar pradesh State
```

```python
sales_state=df.groupby(['State'],as_index= False)['Amount'].sum().sort_values(by='Amount',ascending= False).head(8)
ax = sns.barplot(x='State',y='Amount',data =sales_state,hue="State",palette="husl")
sns.set(rc={'figure.figsize':(15,10)})


```

```python
#also we can see that from above plot people from Uttar Pradesh , spent the most.
```

```python
df['Marriage'].unique()
```

```python
ax = sns.countplot(x='Marriage',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
#sns.set(rc={'figure.figsize':(7,5)})
```

```python
#from the above plot , we can see that count of unmarried people is more in this dataset
```

```python
sales_state=df.groupby(['Marriage','Gender'],as_index= False)['Amount'].sum().sort_values(by='Amount',ascending= False).head(8)
ax = sns.barplot(x='Marriage',y='Amount',data =sales_state,hue="Gender")
sns.set(rc={'figure.figsize':(15,10)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#from the above plot we can see that unmarried females spent the most , whereas married male spent the least.
```

```python
ax = sns.countplot(data=df,x='Occupation')
for bars in ax.containers:
    ax.bar_label(bars)
sns.set(rc={'figure.figsize':(25,5)})
```

```python
sales_occupation=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(data=sales_occupation,x='Occupation',y='Amount')
```

```python
# from the above plot , we can see that the IT sector people spent the most; whereas people from Agriculture sector spent the least
```

```python
ax=sns.countplot(data=df,x='Product_Category')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# from the above plot , we can see that Clothing and Apparel category dominates the dataset
```

```python
heatmap_data = pd.crosstab(df['Occupation'], df['Zone'])

plt.figure(figsize=(10,6))
sns.heatmap(heatmap_data, annot=True, fmt="d", cmap="crest")

plt.title("Occupation vs Zone")
plt.xlabel("Zone")
plt.ylabel("Occupation")
plt.show()
```

```python
#from the above heatmap , we can see that IT sector dominates in all the zones , with agiriculture being the least dominating
```

```python
num_cols = ['Age', 'Orders', 'Amount']

# Correlation matrix
corr = df[num_cols].corr()

plt.figure(figsize=(6,4))
sns.heatmap(corr, annot=True, cmap="coolwarm", center=0, fmt=".2f")

plt.title("Correlation Heatmap (Age, Orders, Amount)")
plt.show()
```
